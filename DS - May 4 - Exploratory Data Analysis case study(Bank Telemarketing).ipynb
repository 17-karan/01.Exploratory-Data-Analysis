### Making Buckets from Age column

```python
data['Age Group'] = pd.cut(data.age,[0,30,40,50,60,100],labels=["<30","30-40","40-50","50-60","60+"])
data[['age','Age Group']]
```

```python
age_grp = data.groupby('Age Group')
age_grp['response_flag'].mean()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
data['Age Group'].value_counts(normalize=True).plot.bar()
plt.xticks(color='red',fontweight='bold')
plt.yticks(color='red',fontweight='bold')
plt.xlabel('Age Group',fontdict={'fontweight':'bold','fontsize':12,'color':'blue'})
plt.subplot(1,2,2)
data.groupby('Age Group')['response_flag'].mean().plot.bar()
plt.xticks(color='red',fontweight='bold')
plt.yticks(color='red',fontweight='bold')
plt.xlabel('Age Group',fontdict={'fontweight':'bold','fontsize':12,'color':'blue'})
plt.show()
```

```python
plt.figure(figsize=(18,9))
data.groupby('job')['response_flag'].mean().plot.bar()
plt.xticks(color='red',fontweight='bold',rotation=1,fontsize=12)
plt.yticks(color='red',fontweight='bold',fontsize=12)
plt.xlabel('Job',fontdict={'fontweight':'bold','fontsize':16,'color':'blue'})
plt.show()
```
